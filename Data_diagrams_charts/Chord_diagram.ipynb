```python
import pandas as pd
import numpy as np
from itertools import chain
from PIL import Image

import holoviews as hv
from holoviews import opts, dim
from holoviews.plotting.bokeh.styles import (line_properties, fill_properties, text_properties)
from bokeh.sampledata.airport_routes import routes, airports
hv.extension('bokeh')
```

```python
data = pd.read_csv('C:/Users/Ceiec04/Desktop/Prediccion_Alertas_Alimentarias/Datasets/full_RASFF_DATA_16092019.csv', sep=';', 
                   header=0, index_col = 0)
data['DATE_CASE'] =pd.to_datetime(data.DATE_CASE, yearfirst=True, format = '%d/%m/%Y')
```

# Preprocesamiento

```python
origen_not = []
rel_not = []
dest_not = []
ori_dest_not = []
ori_rel_not = []
dest_rel_not = []
route_dist = []
route_dest = []

for index, row in data.iterrows():
    if row['NOT_COUNTRY'] in row['COUNT_ORIGEN'] and row['NOT_COUNTRY'] in row['COUNT_DESTIN']:
        ori_dest_not.append(row['NOT_COUNTRY'])
    if row['NOT_COUNTRY'] in row['COUNT_ORIGEN'] and row['NOT_COUNTRY'] not in row['COUNT_DESTIN']:
        origen_not.append(row['NOT_COUNTRY'])
    if row['NOT_COUNTRY'] in row['COUNT_DESTIN'] and row['NOT_COUNTRY'] not in row['COUNT_ORIGEN']:
        dest_not.append(row['NOT_COUNTRY'])
        if row['NOT_COUNTRY'] == 'Italy' or row['NOT_COUNTRY'] == 'United Kingdom' or row['NOT_COUNTRY'] == 'Germany' or row['NOT_COUNTRY'] == 'Denmark' or row['NOT_COUNTRY'] == 'France':
            origen=row['COUNT_ORIGEN'].split(",")  
            for o in origen:
                route = []
                route.append(row['NOT_COUNTRY'])
                route.append(o)
                route_dest.append(route)                       
    if row['NOT_COUNTRY'] in row['COUNT_CONCERN']:
        rel_not.append(row['NOT_COUNTRY'])
        if row['NOT_COUNTRY'] == 'Italy' or row['NOT_COUNTRY'] == 'United Kingdom' or row['NOT_COUNTRY'] == 'Germany' or row['NOT_COUNTRY'] == 'Spain' or row['NOT_COUNTRY'] == 'Netherlands':
            origen=row['COUNT_ORIGEN'].split(",")  
            for o in origen:
                route = []
                route.append(row['NOT_COUNTRY'])
                route.append(o)
                route_dist.append(route)
    if row['NOT_COUNTRY'] in row['COUNT_CONCERN'] and row['NOT_COUNTRY'] in row['COUNT_ORIGEN']:
        ori_rel_not.append(row['NOT_COUNTRY'])
    if row['NOT_COUNTRY'] in row['COUNT_CONCERN'] and row['NOT_COUNTRY'] in row['COUNT_DESTIN']:
        dest_rel_not.append(row['NOT_COUNTRY'])   
```

```python
routes = pd.DataFrame(route_dest)
routes.columns = 'Notificante', 'Origen'
```

```python
Italia = routes[routes['Notificante'].str.contains("Italy")]
Italia = Italia.groupby(['Notificante','Origen']).size().reset_index().rename(columns={0:'count'})
Italia = Italia.sort_values(by=['count'], ascending = False)
Italia = Italia[:5].reset_index(drop = True)
Italia
```

```python
United_Kingdom = routes[routes['Notificante'].str.contains("United Kingdom")]
United_Kingdom = United_Kingdom.groupby(['Notificante','Origen']).size().reset_index().rename(columns={0:'count'})
United_Kingdom = United_Kingdom.sort_values(by=['count'], ascending = False)
United_Kingdom = United_Kingdom[:5].reset_index(drop = True)
United_Kingdom
```

```python
Germany = routes[routes['Notificante'].str.contains("Germany")]
Germany = Germany.groupby(['Notificante','Origen']).size().reset_index().rename(columns={0:'count'})
Germany = Germany.sort_values(by=['count'], ascending = False)
Germany = Germany[:5].reset_index(drop = True)
Germany
```

```python
Denmark = routes[routes['Notificante'].str.contains("Denmark")]
Denmark = Denmark.groupby(['Notificante','Origen']).size().reset_index().rename(columns={0:'count'})
Denmark = Denmark.sort_values(by=['count'], ascending = False)
Denmark = Denmark[:5].reset_index(drop = True)
Denmark
```

```python
France = routes[routes['Notificante'].str.contains("France")]
France = France.groupby(['Notificante','Origen']).size().reset_index().rename(columns={0:'count'})
France = France.sort_values(by=['count'], ascending = False)
France = France[:5].reset_index(drop = True)
France
```

```python
frames = [Italia, United_Kingdom, Germany, Denmark,France]
```

```python
alerts_routes = pd.concat(frames)
alerts_routes_count = alerts_routes.reset_index(drop=True)
alerts_routes_count
```

```python
alert_nodes = alerts_routes_count['Notificante'].unique().tolist()
alert_nodes. append(alerts_routes_count['Origen'].unique().tolist())
```

```python
def flatten(A):
    rt = []
    for i in A:
        if isinstance(i,list): rt.extend(flatten(i))
        else: rt.append(i)
    return rt
```

```python
alert_nodes = flatten(alert_nodes)
alert_nodes = alert_nodes
alert_nodes = list(set(alert_nodes))
```

```python
def swapPositions(list, pos1, pos2): 
      
    list[pos1], list[pos2] = list[pos2], list[pos1] 
    return list
```

```python
Italy_ix = alert_nodes.index("Italy")
alert_nodes = swapPositions(alert_nodes, Italy_ix, 0)
United_Kingdom_ix = alert_nodes.index("United Kingdom")
alert_nodes = swapPositions(alert_nodes, United_Kingdom_ix, 1)
Germany_ix = alert_nodes.index("Germany")
alert_nodes = swapPositions(alert_nodes, Germany_ix, 2)
Denmark_ix = alert_nodes.index("Denmark")
alert_nodes = swapPositions(alert_nodes, Denmark_ix, 3)
France_ix = alert_nodes.index("France")
alert_nodes = swapPositions(alert_nodes, France_ix, 4)
```

```python
alert_nodes = pd.DataFrame(alert_nodes)
alert_nodes.columns = ['Paises']
alert_nodes
```

# Diagrama de cuerdas (top 5 notificantes y distribuidores)
